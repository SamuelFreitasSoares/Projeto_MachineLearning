```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

df = pd.read_csv(r'..\dados\dataset_pronto_para_modelagem.csv')

df['vegetacao_prevista'] = df['vegetacao_natural'] * np.random.uniform(0.95, 1.05, len(df))

```

```python
# üìä 3. Visualiza√ß√£o de insights da AED
print("Informa√ß√µes b√°sicas do dataset:")
display(df.describe())

plt.figure(figsize=(10,6))
sns.histplot(df['vegetacao_natural'], kde=True)
plt.title("Distribui√ß√£o da Vegeta√ß√£o Natural")
plt.xlabel("√Årea de vegeta√ß√£o natural")
plt.ylabel("Frequ√™ncia")
plt.show()

plt.figure(figsize=(10,6))
sns.countplot(x='bioma', data=df)
plt.title("Quantidade de registros por Bioma")
plt.xticks(rotation=45)
plt.show()

# -----------------------------------

# ü§ñ 4. Visualiza√ß√£o dos resultados do modelo
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['vegetacao_natural'], y=df['vegetacao_prevista'], alpha=0.6)
plt.title("Compara√ß√£o entre valores reais e previstos")
plt.xlabel("Vegeta√ß√£o Natural (Real)")
plt.ylabel("Vegeta√ß√£o Prevista (Modelo)")
plt.show()

r2 = r2_score(df['vegetacao_natural'], df['vegetacao_prevista'])
rmse = mean_squared_error(df['vegetacao_natural'], df['vegetacao_prevista'], squared=False)

print(f"Desempenho do Modelo:")
print(f"R¬≤ = {r2:.3f}")
print(f"RMSE = {rmse:.3f}")
```
